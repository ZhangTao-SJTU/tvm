```python
from toolbox import *


initial_time=0
cell_division_time=100
final_time=200
time_step=0.005
dump_vtk=10
log=10
s0=5.6
Lth=0.02
T=1e-4
kv=10
box_x=4
box_y=4
box_z=4
periodic_x='p'
periodic_y='p'
periodic_z='p'

build_dir = 'build/'
if not os.path.isdir(build_dir):    
    os.mkdir(build_dir)




def compile():
    os.system('cd {} && cmake ../ && make'.format(build_dir))
    return
#create_conf_file()

def make_conf(t_init=0,t_final=100000):
    with open(build_dir+'conf', 'w') as f:
        f.write('time {} {} {}\n'.format(t_init,t_final,time_step))
        f.write('dump vtk {}\n'.format(dump_vtk))
        f.write('log {}\n'.format(log))
        f.write('s0 {}\n'.format(s0))
        f.write('Lth {}\n'.format(Lth))
        f.write('T {}\n'.format(T))
        f.write('kv {}\n'.format(kv))
        f.write('box {} {} {} {} {} {}'.format(box_x,box_y,box_z,periodic_x,periodic_y,periodic_z))
    return
def set_initial_conditions(t_init=0,t_final=100):
    #create conf file
    make_conf(t_init,t_final)
    #create sample.topo
    os.system('cd {} && python3 ../scripts/tvm/main.py'.format(build_dir))
    return



def run_simulation():
    compile()
    set_initial_conditions(initial_time,cell_division_time)
    os.system('cd {} && ./tvm'.format(build_dir))
    sample=periodic_box(time=cell_division_time)
    # Insert cell division here
    
    make_sample_dot_topo(build_dir,sample.vertices_,sample.edges_,sample.polygons_,sample.cells_)
    #make_conf(cell_division_time+dump_vtk,final_time)
    #os.system('cd {} && ./tvm'.format(build_dir))
    
    return

```

```python
run_simulation()
```
